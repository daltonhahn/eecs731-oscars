## Data Visualization and Storytelling

### Word Cloud of most used words in dataset

#### Inspiration from https://www.datacamp.com/community/tutorials/wordcloud-python

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import math
from statistics import mean, stdev

from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv("../data/external/Shakespeare_data.csv")
```

```python
df.head()
```

```python
# Remove NaN rows from the dataset (these represent stage directions/non-dialogue)
print("With NaNs = ", df.shape)
df = df.dropna()
print("Without NaNs = ", df.shape)
df = df.reset_index(drop=True)
df.head()
```

```python
# Start with one review:
text = " ".join(line for line in df.PlayerLine)

stopwords = set(STOPWORDS)
stopwords.update(["will", "thy", "thee"])
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords, max_font_size=50, max_words=100, background_color="white").generate(text)

# Display the generated image:
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
